## `Pydantic` para crear Entidades en vez de usar `def(self)` y su función constructora

- `pydantic` usa **metaprogramación** para generar automaticamente:
  * constructor `__init__`
  * validación de tipos `hint type`
  * conversión de datos (coercion de `str` a `int`)
  * implementa transformacion inclusive de listas a JSON automaticamente sin tener que deifninir un metodo dentro de una clase normal para retornarlo asi
  * No te preocupas por el orden de argumentos


````python

from pydantic import BaseModel

class User(BaseModel):
    name: str #le indicamos tipos de datos
    age: int

nuevo_user = User(name="Simon", age="22")

user_list = []
user_list.append(nuevo_user)

user_specific = [
    user_data
    for user_data
    in user_list
    if( user_data.age == 22)

][0]
print(user_specific)
print(nuevo_user)
````

````python
import json
class User():
    def __init__(self, name: str, age: int):
        self.name = name
        try: #validación de tipo str a int
            self.age = int(age)
        except:
            raise "age should be of type integer"
    def return_obj_to_JSON(self):
        return json.dumps(
            {
                "name": self.name,
                "age": self.age,
            }
        )
    
nuevo_user = User(name="Moore", age="22")
print(nuevo_user.return_obj_to_JSON())
print(f"name={nuevo_user.name} age={nuevo_user.age}")
print(
    User("Juan","22").return_obj_to_JSON()
) #instanciamos la clase y retornamos el objeto en JSON
print(
    type(
        User("Juan","22").return_obj_to_JSON()
    ) #instanciamos la clase y retornamos el objeto en JSON
)
````
