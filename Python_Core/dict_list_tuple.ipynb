## 1 - dict vs list (con mentalidad para backend)
- un JSON se ve asi:
```json
{
    "id": 1, 
    "name": "Ana",
    "active": true
}
```

- En `Python`, se ve asi:
```python
user = {
    "id": 1, 
    "name": "Ana",
    "active": True
}
```
- Obtengo esto `Ana` cuando hago esto en **Python** `user["name"]`

````python
user = {
    "id": 1, 
    "name": "Ana",
    "active": True
}
print(user["name"])
````

## 2 - List comprehension (version backend)
- si se tiene esto y solo quiero los miembros activos:
```python
users = [
    { "name": "Ana", "active": False },
    { "name": "Xavier", "active": True },
    { "name": "Sam", "active": True },
]
```
    - ¿Cómo lo escribirías usando list comprehension?

````python
users = [
    { "name": "Ana", "active": False },
    { "name": "Xavier", "active": True },
    { "name": "Sam", "active": True },
]

active_users = [
    user
    for user
    in users
    if user["active"]
]

print(active_users)
````

## Mini Reto
Supón que ahora tienes:
``
users= [
    { "name": "Ana", "age": 22, "active": False },
    { "name": "Xavier", "age": 26, "active": True },
    { "name": "Sam", "age": 21, "active": True },
    { "name": "Laura", "age": 30, "active": True },
]
``


````python
users= [
    { "name": "Ana", "age": 22, "active": False },
    { "name": "Xavier", "age": 26, "active": True },
    { "name": "Sam", "age": 21, "active": True },
    { "name": "Laura", "age": 30, "active": True },
]

active_over_23_years_old_age_unordered = [
    user
    for user 
    in users
    if user["active"] and user["age"] > 23
]

active_over_23_years_old_age_ordered = sorted(
    active_over_23_years_old_age_unordered, key=lambda u:["age"], reverse=True
)

print(active_over_23_years_old_age_ordered)
````

### Consejo de funciones `lambda` 👁️ - es como una funcion anonima de JS
- En JS:
    ```js
    (u) => u.age
    ```
- En Python:
    ```python
    lambda u: u["age"]
    ```
    - porque en Python inetrnamente hace algo como:
    `criterio = key( u )`
    - y luego ordena usando ese criterio .
    `key= lambda u: u["age"]`
