# Manejo de Errores en Python

  - Exisen multiples errores que pasen en backend todo el tiempo:
    - payload incompleto
    - tipo de incorrecto
    - conexión a la DB falla
    - División por cero
    - archivo no existente

    - Para ello `Python` usa **excepciones**
  
## Estructura básica

```python
try:
    codigo_peligroso()
except ErrorTipo:
    manejar_error()
```
- Ejemplo simple:

````python
try:
    x = 10/0
except ZeroDivisionError:
    print("No puedes dividir entre cero")
````

## Capturar el error como variable
- utiliza en la sentencia del `except` + `ValueError` + la keyword `as` + `nombre_variable`, ejemplo:

````python
try:
    x = int("hola")
except ValueError as e:
    print(e)
````

## Multiples errores
- usaras un solo `try`, seguido de su comportamiento a ejecutar si es valido
- junto con diversos `except` que captaran diversos errores, ejemplo:

````python
try:
    value = int(input())
except ValueError:
    print("no es un numero")
except TypeError:
    print("tipo incorrecto")
````

## `else` como manejo de errores junto con `try`
- es poco conocido, y se ejecuta cuando **no hubo error**, se puede decir que va encadenado, ejemplo: 


````python
try:
    x = int("10")
except ValueError:
    print("error")
else:
    print("todo bien", x)
````

## `finally` – siempre se ejecuta
- muy usado para cerrar recursos

````python
try:
    file = open("data.txt")
except FileNotFoundError:
    print("el archivo no existe")
finally:
    print("cerrando proceso")
````

## `raise` Lanzar errores manualmente
- Es muy usado en BackEnd para manejar errores, ejemplo

````python
age = -5

if age < 0:
    raise ValueError("age cannot be negative")
````

## Patrón backend típico
- Ejemplo de validación de un payload:

````python
user = {
    "NAME": "Luis",
    "age": "23"
}
def validate_user(data: dict):
    try: 
        name = data["name"]
        age = int(data["age"])
    except KeyError:
        raise ValueError("Llave faltante / missing Key!")
    except ValueError:
        raise ValueError("La edad debe de ser entero / age must be integer")
    
    return name, age

validate_user(user)
````

## Mini ejercicio 1:
- Corrige esta función para manejar errores:
    ```python
    def divide(a, b):
    return a / b
    ```
    - Queremos:
        - si `b == 0` -> error manejado
        - si `a` o `b` no son numeros -> error manejado
    
    - Escribe la funcion usando `try` `except`

````python
def divide(a: int, b: int) -> int:
    result = 0
    try:
        result = a / b
        return result
    except ValueError:
        raise ValueError("b cannot be 0")

    
divide(3, 0)
````

## 🔥 Correcion y observaciones de mini ejercicio 1

```python
def divide(a: int, b: int) -> int:
    result = 0
    try:
        result = a / b
        return result
    except ValueError:
        raise ValueError("b cannot be 0")
```
  - **Problemas importantes**:
   1) Error al dividir entre cero *NO* es `ValueError`
       - Socrátoc: si haces `10 / 0`, ¿Python se queja por "valor invalido" o por "división entre cero"?
       - ✅ Python lanza: `ZeroDivisionError`, no `ValueError` 
       - Entonces el `except` `ValueError`: nunca atrapará ese caso.
    2) El type hint de retorno no coincide:
        - tu firma dice: `-> int`, pero `a / b` en Python normalmente devuelve **float** (aunque ambos sean `int`)
        - Ejemplo conceptual:
          - `5 / 2` ➡️  `2.5` ***(float)***
        - Entonces debería ser: `-> float` 
    3) Falta el caso de ***"No numeros"***
       - Si te pasan `"10"` y `"2"`, puede fallar como TypeError (dependiendo). 

````python
def divide(a: float = None, b: float = None) -> float:
    try:
        return print(a / b)
    except ZeroDivisionError as e:
        print(e)
    except TypeError as e: #atrapa cuando no pasamos parametros ya que los asigna por defecto como None type value
        print(f"{e} -> {a}:{type(a)} -> {b}:{type(b)}")


divide(3, 0)

divide("3",0)

divide("3", "3")
divide()
````
