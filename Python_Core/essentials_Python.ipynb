## 10 List comprehension (crear listas de forma concisa)
- puedes aplicar dentro de ellas:
    - operaciones
    - filtros a elementos existentes
- Estructura básica:
    ``
    [expresion for elemento in iterable]
    [expresion for elemento in iterable if condicion]
    ``
- en JS para `map`
    ```js
    let numeros = [1, 2, 3, 4]
    let numerosAlCuadrado = numeros.map( n => n ** 2)
    console.log(numerosAlCuadrado); //[1, 4, 9, 16]
    ```
- en JS para `.filter`:
    ```js
    let numeros = [1, 2, 3, 4]
    let pares = numero.filter( n => n%2 == 0 )
    /// --- o tambien:
    let numeros = [1, 2, 3, 4]
    let pares = numeros.filter(function (n) {
        return n%2 == 0
    })
    console.log(pares) // [2, 4]
    ```

````python
numeros = [1, 2, 3, 4, 5]
numerosAlCuadrado = [n ** 2 for n in numeros]
print("Numeros al cuadrado: ",numerosAlCuadrado)
numeros[2] *= 3
print("sin modificacion a lista hija: ",numerosAlCuadrado)

pares = [ n for n in numeros if (n%2 == 0)] 
print("Pares: ",pares)
````

## Mini Reto Rápido
1. Crear una lista de usuarios (diccionarios) con:
    - name
    - age
    - active
2. Filtra solo los activos
3. Devuelve solo los nombre usnado list comprehension.


````python
habitantes_Loft = {
    "0": {
        "name": "Ana",
        "age": 22,
        "active": False
    },
    "1": {
        "name": "Xavier",
        "age": 26,
        "active": True
    },
    "2": {
        "name": "Sam",
        "age": 21,
        "active": True
    },
    "3": {
        "name": "Turista",
        "age": 28,
        "active": True
    }
}
print(habitantes_Loft)

activos_loft = [
    user for user in habitantes_Loft.values() if user["active"] == True
]
print(activos_loft)
````

````python
print(habitantes_Loft.keys())

active = [ 
    user["name"]                        # Dame (este value de correspondiente a este key)
    for user                            # Para cada (obj anidado/hijo)
    in habitantes_Loft.values()         # en (este objeto padre)
    if user["active"]                   # Si (obj[llave /key] == condicion)
]
print(active) 
````
