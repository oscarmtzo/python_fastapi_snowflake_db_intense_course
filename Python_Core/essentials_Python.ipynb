## 1) Variables en Python (tipado dinámico)

- Python infiere el tipo automáticamente.

````python
x = 5  # x es un entero
print(x)  # Imprime: 5

age = 30  # int
price = 10.5  # float
name = "Turista"  # str
is_active = True  # bool
value = None  # equivalente a null en JS, es falsey

print(type(age), type(price), type(name), type(is_active), value)
````

## 3) Diccionarios (como objetos en JS)

En JavaScript:

```js
let user = {
  name: "Ana",
  age: 30
}
```

En Python:

````python
numbers = [1, 2, 3]
names = ["Ana", "Luis"]
print(names[0])

user = {
    "name": "Ana",
    "age": 30
}

print(user)
print(user["name"])  # acceso recomendado con corchetes
````

## 5) Loops

En JavaScript:

```js

```

````python
if age > 18:
    print("Adulto", age)

for number in numbers:
    print(number)
````

## 7) Type hints

- No es obligatorio, pero FastAPI lo usa mucho.

````python
def add(a, b):
    return a + b

print(add(2, 3))


def add_typed(a: int, b: int) -> int:
    return a + b

print(add_typed(10, 20))
````
