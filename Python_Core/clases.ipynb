# Anatomia de una Clase 
```python
class User:
    def __init__(self, name: str, age:int):
        self.name = name
        self.age = age

```
- `self` es similar a `this` en el contexto de ser un parametro que se paras a la función constructora `__init__` y siempre es el primer argumento de dicha función 
  - Permite que cad ainstancia de esta clase, pueda tener sus propios datos, independientes a los que pudieran tener otro objetos instanciados de la misma clase

## Mini Reto 1 - Crear clase completa:
- Escribir una clase Car con:
  - atributos: `brand` (str), ``year`` (int)

````python
class Car:
    def __init__(self, brand: str, year: int):
        if (brand):
            self.brand = brand
        else: raise TypeError

        if (year < 2026):
            self.year = year
        else: raise ValueError("year cannot be in the future")
        self.year = year

    def printBrand(self) -> str:
        return self.brand
    
    def car_info(self) -> str:
        #return self.brand + ' - ' + str(self.year)
        return f"{self.brand} - {self.year}"
        
    def is_classic(self) -> bool:
        if((2025 - self.year) > 25) : return True
        else: return False

dodge_charger = Car("dodge", 2020)
dodge_charger.brand
dodge_charger.car_info()
dodge_charger.is_classic()
````
