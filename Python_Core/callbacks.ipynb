# Callbacks

- En JavaScript:

````python
%%script node 
function printData(data){
    console.log(data)
} 
function process(data, callback) {
    callback(data)

}
process("Hello world", printData)
````

````python
def process(data: str, callback) -> str:
    return callback(data)

def printData(data: str) -> str:
    return data

process("hello world", printData)
````

- Ejemplo más real:

````python
def apply_operation(x: int, func) -> int:
    try:
        return func(x)
    except TypeError as e:
        print("\t",e)

def square(n: int) -> int:
    try:
        return n * n
    except TypeError as e:
        print(f"{"\t"} -> {e}")

apply_operation("3", square)
apply_operation(3, square)
````

## 🎯 Mini ejercicio 2 (callbacks)
- Escribe una función:
    ```python
    def apply(numbers, func):
    ```
    - que aplicque una función a cada número. 
  - Ejemplo esperado:
        ```python
        apply([1, 2, 3], lambda x: x*2)
        ```

````python
print(f"Hand verification for data types type(2) == type(1): {type(2) == type(1)}")

def apply(nums: list[int], callback) -> list[int]:
    try:
        print(f"Your input list is: {nums}")
        return callback(nums)
    except TypeError as e:
        print(e)

def double_num(nums: list[int]) -> int:
    try:
        return [
                num * 2
                for num
                in nums
                if (type(num) == type(1))
            ]
    except TypeError as e:
        print(e)

print(
    apply(["2", 2, 3], double_num)
)

print(
    apply([1, 2, 3], double_num)
)
````

## 🔥 Correcion y observaciones de mini ejercicio 2
- Es buena idea usar validaciones, pero aqui hay un detalle conceptual clave:
  - **El apply no applica el callback a cada elemento** ya que tu `apply` hace:
    ```python
    return callback(nums)
    ```
    - o sea: le pasa toda la lista al callback, y el ejercicio decia: *aplicar una fucnión a cada número*, Entonce `apply` debería iterar y ejecutar el callback por elemento
    - 

````python

def apply(nums: list[float] = None, callback = None) -> float:
    try:
        if (len(nums) > 0):
            return print(
                [
                    callback(num)
                    if(type(num) == type(1))
                    else raise_error_when_str()
                    for num 
                    in nums
                ]
            )
        else:
            return print("No elements inside the list")
    except TypeError as e:
        print(e) # maneja el error junto con None si no le paso parametros a la funcion

def double_number(num):
    try:
        return num * 2
    except TypeError as e:
        print(e)

def raise_error_when_str():
    ## returning a str to not break the code, altough it might not be correct, just to keep next code flowing 
    return ValueError("String exists within your list")


apply([2,"3"], double_number)
apply(["2","3"], double_number)
apply([], double_number)
apply([], )
apply([1])
apply([2,3], double_number)
````
