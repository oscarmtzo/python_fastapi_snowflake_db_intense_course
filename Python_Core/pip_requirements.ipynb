# `venv` – ambiente virtual de python
    ```
    🧠 Resumen:
    1 crear venv
    2 activar venv
    3 instalar paquetes
    ```
- es especialmente útil cuando no quieres romper otras aplicaiones a nivel general (global) en tu PC al actualizar librerias a una version especifica debido a los requerimientos de la aplicación
- Se instala de esta forma en la terminal de comandos:
    ```sh
    python3 -m venv venv
    ```
    - esto crea una carpeta llamada: 
        ```
        venv/
        ```
        - dentro de esta carpeta `Python` guarda:
            - su propio **interprete**
            - sus propias **libraries**
            - su propio `pip`
    - Es como un **Python aislado solo para el proyecto**
    ## Activar el ambiente virtual
    - En Mac/Linux:
        ```sh
        source venv/bin/activate
        ```
        - Cuando se activa, verás algo asi en la terminal:
            ```
            (venv) <User>@<host> proyecto %
            ```
    - En Windows:
        ```bash
        venv\Scripts\activate
        ```
    ## Desactivar el entorno virtual:
    - `Mac` / `Linux` / `Windows`
        ```bash
        deactivate
        ```

        ### Instalar paquetes dentro del ambiente
        - Cualquier instalación se queda **solo en ese proyecto**
            ```bash
            pip install fastapi
            pip install uvicorn
            ```
    ## Salir del ambiente
    - Usa este comando para salir del ambiente:
        ```bash
        deactivate
        ```

    ## Estructura típica de un proyecto
    ```txt
    mi_proyecto/
    │
    ├── venv/
    │
    ├── main.py
    │
    └── requirements.txt
    ```

    ## Guardar dependencias 
    - si en el proyecto utilias bibliotecas o frameworks de código, es necesario que mantengas seguimiento de las mismas ya que al momento de compartir tu desarrollo por `GitHub`, cuando tus compañeros descarguen tu proyecto y necesiten continuar el desarrollo, agiliza el tener las dependencias en el archivo `requirements.txt`
      - `pip freeze > requirements.tx` agiliza este proceso, ya no tienes que escirbirlas a mano.
  
        ```txt
        fastapi==0.110.0
        uvicorn==0.29.0
        ```

    - otra persona puede instalar todo con:
        ```bash
        pip install -r requirements.txt
        ```

# `pip` y `requirements.txt` ⩬ `npm` y `package.json`
- `pip` es el gestor de paquetes (nos descarga librerias desde la interfaz de linea de comandos)
- `requirements.txt` es la lista de bibliotecas, fameworks -> **dependencias** que ocupa nuestro código ademas de las que ya trae por defecto `python` 
- Se conecta a [Python Package Index](https://pypi.org), dodne podemos encontrar bibliotecas y frameworks, como:
  - `fastapi`
  - `uvicorn`
  - `pydantic`
  - `requests`
  - `numpy`
- Cuando ejecutas:
    ```bash
    pip install fastapi
    ```
    `pip` hace esto:
    ```
    PyPI
        ↓
    descargar paquete
        ↓
    resolver dependencias
        ↓
    instalar en site-packages
    ```


- si instalas usas un ambiente virtual con python `venv`, las dependencias estrán instaladas en la siguiente ruta:
  ```
  project/
    venv/
        lib/pythonX/site-packages/
  ```

## Comandos de `pip` que usa un backend dev



|comando|para qué sirve|
|---|---|   
| `pip install <paquete>` | instalar nuevos paquetes|
| `pip uninstall <paquete>` | remover un paquete instalado|
| `pip list` | ver paquetes instalados |
| `pip show <paquete>` | informacion de un paquete |
| `pip freeze ` | lista exacta de versiones |
| `pip freeze > requirements.txt` | lista exacta de versiones y el resultado los ingresa al archivo definido mediante piping usando operadores de `bash` |
| `pip install -r requirements.txt` | instalar dependencias definidas en ese archivo|

## Ejercicio 1 – inspeccionar un paquete
- instalar algo pequeño como: `pip install requests` [requests](https://pypi.org/project/requests/), e inspeccionalo:
    
    ```bash
    pip show results
    ```
    **Resultado:**
    ```bash
    Name: requests
    Version: 2.32
    Location: .../site-packages
    Requires: charset-normalizer, urllib3
    ```
    - `requests` necesita de los paquetes definidos en `Requires: charset-normalizer, urllib3`, por si solo no es suficiente pero ya los trae consigo, por eso cuando lo instalas, y creas el `requirements.txt` tmabien incluye dichos paquetes, mira:
        ```bash
        certifi==2026.2.25
        charset-normalizer==3.4.4
        idna==3.11
        requests==2.32.5
        urllib3==2.6.3

        ```

## Ejercicio 4 – reproducir entorno
- Simula que alguien clona tu repo
1) Borra el `.venv` – entorno virtual
2) Crea uno nuevo:
    ```bash
    python -m venv venv # crear ambiente virtual llamada venv, la ultima opcion es el nombre de la carpeta
    source venv/ #activar el ambiente virtual
    ```

    - `python3` en el caso de tener versiones 2 y 3 de python
    - la opción `-m` indica a python que use el siguiente texto (despues del espacio) como argumento de busqueda para que encuentre el modulo `.py`con esa coincidencia y lo ejecute como un ***script***
    
    
   - puedes usar `rm -rdf` para borrar desde linea de comandos Unix archivos, caprteas anidados de manera recursiva sin pedir conrifmaciones en la `cli` para cada archivo o carpeta

3) Instalar dependencias:
    ```
    pip install -r requirements.txt 
    ``` 

    - con esto tu entorno será **identico al original**

# Regla de **oro** que usan los equipos para instalar dependencias

- Cuando trabajas en un proyecto, e instalas dependencias, una buena practica es usar:
```bash
pip install -r requirements.txt
```

- Para agregar una nueva dependencia:
```bash
pip install <nueva_dependencia>
pip freeze > requirements.txt #actualiza la refernecia escrita de dependencias 
```

- ya que si solo actualizas con `pip install <paquete_ya_instalado>` te descaragrá la pocion mas nueva y eso puede romper tu entorno.
