# Funciones con Type Hints
- en `JS` se definenn de esta manera:
    ```js
    function add(a, b) {
        return a + b;
    }
    ```

- En `Python` es de esta manera:
    ```python
    def add(a, b):
        return a + b
    ```

## ¿Qué es un **type hint**?
- es una nota que le pones a la funcino para que als personas y herramientas sepan que esperas.
    ```python
    def add(a: int, b: int) -> int:
        return a + b
    ```
    - esto retorna un entero, la definicion de la funcion es parecida a las `arrow functions en JS`
    ```js
    let add = (a, b) => a + b;
    ```
- ¿Si le paso strings a esa función se romperia?
    - No necesariamente, *Type Hints* es como una documentación viva
    - No realiza ``cast``, **No convierte nada**, solo es promesa/documentacion 

### ¿Para qué sirven en FastAPI?
1. Autocomplete y revisión:
    - Tu editor te ayuda
    - MyPy/linters pueden detectar errores 
2. En FastAPI: se vuelven parte del contrato del API
    - FastAPI usa type hints + Pydantic para:
        - validar requests
            - FastAPI verifica en el caso de que el servidor reciba un `item_id: int` que no llegue de otro tipo como por ejemplo cadena y arrojaria un error:
            ```python
            from fastapi import FastAPI
            app = FastAPI()
            @app.get("/items/{items_id}")
            def read_item(item_id: int):
                return {"item_id": item_id}
            ```
            - si alguien manda ``GET /item/5`` **Funciona**
            - Pero si manda ```GET /items/hola``, FastAPI responde automaticamente y tú **no necesitas escribir alguna validación**:
                ```json
                {
                    "detail": [
                        {
                            "msg": "value is not a valid integer"
                        }
                    ]
                }
                ```
        - documentar Swagger del endpoint
        - Validacion de request bodies con Pydantic:
            ```python
            from pydantic import BaseModel

            class Item(BaseModel):
                name: str
                price: float
            ```
            - En el Endpoint:
                ```python
                @app.post("/items")
                def create_item(item: Item):
                    return item
                ```
            - Si alguine manda:
                ```json
                {
                    "name": "Laptop",
                    "price": "1000"
                }
                ```
            - FastAPI convierte "1000" -> "1000.0"
            - Pero si alguien manda:
                ```json
                {
                "name": "Laptop",
                    "price": "hola" 
                }
                ```
            - FastAPI responde con un error automaticamente
## 🤾‍♀️Mini ejemplo con caso de usuarios
- Caso A: **sin type hints**:

````python
users= [
    { "name": "Ana", "age": 22, "active": False },
    { "name": "Xavier", "age": 26, "active": True },
    { "name": "Sam", "age": 21, "active": True },
    { "name": "Laura", "age": 30, "active": True },
]

def get_active_names(users):
    return [
            user["name"] 
            for user 
            in users 
            if user["active"]
        ]
get_active_names(users)
````

- Caso B **con type hints (más pro)**:

````python
def get_active_names(users :list[dict] ) -> list[str]:
    return [
        user["name"]
        for user
        in users
        if user["active"]
    ]

get_active_names(users)
````

## 🔥 Mini Reto 
- Escribir una función con type hints que:
   - Reciba una lista de numeros
   - Devuelva solo los numeros mayores a 10 

````python
numeros = [1, 2, 3, 4, 5]

def only_numbers_above_10(numeros: list[int]) -> list[int]:
    return [
        num
        for num
        in numeros
        if num >= 10
    ]

only_numbers_above_10([11, 9, 15, 30])
````

## Funciones Lambda en Python
- similares a las funciones anonimas de JavaScript y en su uso, ejemplo:
  ```js
  let sumar = (a, b) => a + b
  ```
  ```python
  sumar = lambda a, b: a + b
  ```
  - aunque no tiene return, su uso tipico es como funciones `.filter()`, `.map()` y `sorted()`, ejemplo:
    ```python
    numeros = [1, 2, 3, 4]
    dobles = list( map( lambda x: x * 2, numeros))
    ```
    - estan limitadas a una sola expresión y ella debe de ser corta, tmapoco puedes imprimir dentro de ellas, si puedes mandar a llamar a otra funcion
