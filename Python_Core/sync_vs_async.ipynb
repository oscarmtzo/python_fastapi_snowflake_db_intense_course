# Sync vs Async (corutinas – coroutines)
- las tareas son bloqueantes, espera la finalización de una para empezar la otra en el `event loop`
  - Porque una tarea bloquea a la otra, ejemplo:

````python
import time
def task1():
    print("Task_1 start")
    time.sleep(2) #bloquea el event loop
    print("Task_1 end")

def task2():
    print("Task_2 start")
    time.sleep(2)
    print("Task_2 end")

task1()
task2()
````

- **Analogia simple**
    ```
    hacer café ☕️
    esperar a que hierva ⏱️
    hacer sandwich 🥪
    ```
    - El **cocinero se queda esperando**
      - Esto es `sync`
***
- `Async` (idea conceptual)
  Mientras el café hierve:
  ```
  hacer sandwich 🥙
  lavar platos 🍴
  revisar pedidos 🛍️
  ```

## En Backend
- en un servidor:
  `request` 👉 `consultar DB`
  - Si es sincrono:
    ``
    servidor espera
    no atiende otros requests
    ``

    - Si es **asincrono**:
      ``
      servidor espera DB
      pero atiende otros clientes
      ``

````python
async def hello():
    print("hola")

await hello() # con await nos aseguramos de encolar esta funcion y cuando este lista retornar el resultado apropiado
hello() # retorna un objeto de corutina (tarea pendiente), pero no el resultado
````

````python
import asyncio

async def hello():
    await asyncio.sleep(1) 
    print("hola") 


#asyncio.run(hello())
````

````python
import asyncio

async def a():
    print("A1")
    await asyncio.sleep(0)
    print("A2")

async def b():
    print("B1")
    await asyncio.sleep(0)
    print("B2")

await a()
await b()
````

# Ejecutando dos o mas funciones de manera concurrente (secuenciales al mismo tiempo)
- se utiliza de la biblioteca `asyncio` el metodo `.gather(a() ,b())`, cuyos parametros serán funciones que se ejecuten concurrentemente, es decir que si existe algun tipo de flujo de datos que sea bloqueante, y que tome algun tiempo en realizarse pasará de la primera a la segunda si es que encunetra ese tipo de bloqueo por procesamiento de tiempo
- `async` es solo recomendado para traer datos que tomen tiempo, pero **no se recomeinda para tareas arduas del CPU** como realizar `ML`, `calculos computacionales complejos` etc.

````python
import asyncio

async def task(n):
    print("start", n)
    await asyncio.sleep(1)
    print("end", n)

async def main():
    await asyncio.gather(task(1), task(2)) # se ejecuta concurrentemente

await main()
````
