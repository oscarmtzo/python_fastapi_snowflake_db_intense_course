# Socket en Python
- **router** es el trasnmisor / instrumento en medio / de tu dispositivo (PC, smartphone, etc.)
- `IP` Public IP Address - fisical location on internet 
  - Within local network - local IP address `192.168.1.176` - IPV4
  - Your router can talk to your specific device using the local IP address, meanwhile the public IP Address is held by your Roouter to ping the repources of an external server outside on the internet

1) importar `sockets`
2) importar `threading`
3) definir `PORT` – puerto de entrada o escucha de nuestro servidor `5050`
4) definir `SERVER` – direccion IPV4, direccion local que nos asina el router dentro de la red local
   ```bash
   ifconfig 
   ```
   -    buscar la direccion IPV4 o tmabien ibtenerla por medio del medoto `.gethostbyname(socket,gethostname())` del objeto `socket` que imprtamos
   ```python
   SERVER =  socket.gethostbyname(socket.gethostname()) #-> imprime la IP local asignada 192.168.1.12
   ```
5) Crear una instancia de un servidor abriendo / creadno un nuevo socket, usando el metodo `.socket()` del objeto importado `socket`
    ```python
    server = socket.socket()
    ```
    - debemos pasarle argumentos para especificar: **tipo de conexion IPV4 = AF_INET** y  tipo de socket **SOCK.STREAM**
    ```python
    server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    ```

6) Definir una direccion `ADDRESS` en la que vincularemos `SERVER_LOCALIP` y `PORT`
