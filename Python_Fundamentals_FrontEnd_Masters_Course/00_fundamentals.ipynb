### Funciones con valores por defecto
- Los argumentos de la funcion que tienen valores por defeco deben de ser  escritos al utlimo dentro de los parentesis, de otra forma obtendremos un `syntax error`
- (❌ ♻️) No se debe prestablecer como valor por defecto una `list` vacia, ya que los parametros asignados a las funciones se instancia una vez, y es al inicio de la ejecucion de la función, por lo que no podrás usar nuevamente la funcion sin que se asignen una y los parametros que le pases, es decir permaneceran dentro de la funcion y ya no poadrá ser reusable

````python
def saludo (nombre, saludo= "Hola "):
    return print(f"{saludo} {nombre}")

saludo("Juan")

def agregar_numero(a, b=[]): #instancia los paramtros 1 sola vez al crearse funcion
    b.append(a)
    return print(b)

agregar_numero(1)
agregar_numero(5) #arrastra el primer valor 😭
agregar_numero(6) #arrastra el primer y segundo valor 😭 y no devuelve una nueva lista, no lo convierte en algo reutilizable ♻️

# lo que se debe hacer es asignarle un valor None
def agregar_numero_mejorado(a, b=None):
    if b ==None:
        b = []
    b.append(a)
    return print(b)

agregar_numero_mejorado(2) # ya no arrastra los valores de inicio
agregar_numero_mejorado(3) # ya no arrastra los valores de inicio
agregar_numero_mejorado(88) # ya no arrastra los valores de inicio
````

### `[].sort()`, `sorted([])` , `[].revert()` – Ordenando listas
- **Ordenar de menor a mayor** – puedes hacerlo de 2 maneras:
  - `sorted()` -> no modifica el arreglo original
  - `[34,14].sort()` -> SI modifica el arreglo original -> `[14, 34]`
- **Revertir el orden actual**:
  - `[19, 201, 2].revert()` -> retorna `[2, 201, 19]`

````python
numeros_suerte = [45, 888, 24]

print(
    f"{sorted(numeros_suerte)} "sorted" no modifica el arreglo original \n arreglo original= {numeros_suerte}"
)

numeros_suerte.sort() #este si modifica el nuevo arreglo
print(numeros_suerte)


nuevos_numeros_suerte = [204, 305, 40]
print(nuevos_numeros_suerte)
nuevos_numeros_suerte.reverse()
print(nuevos_numeros_suerte)
````

## `set()` `a={1,2}` – Sets - inmutables - **Alamcenamiento de datos unicos** sin duplicados
- no tiene un orden, no se pueden acceder por indice
- son muy rapidos, alamacenan el `hash value`, es muy rapido de ver si un elemnto existe


````python
# Sets
# {} su tipo es un diccionario
set()#debes ser explicito en crear un set porque aunque el tipo de dato es una 
print(type({}))

print(set())
print(type({1}))

nombres = {"Alice", "Alice", "Bob", "Bob","Charlie", "Diana", "Eduardo", "Fernanda", "Gabriel", "Hannah"} # ignoran los datos duplicados
print(nombres) # No imprime o alamcena los datos duplicados

#Chequemos un has, es la forma de bver de un algoritmo si un elemento existe dentro de un set, no aplica para listas – porque no se puede hashear
try: 
    print(nombres[0])
except Exception as e:
    print(f"{e} -> no existen las posiciones espoecificas ni su acceso desde un set")

#si puedes remover o añadir elementos a un set()
print(nombres) #{"Hannah", "Charlie", "Diana", "Bob", "Fernanda", "Gabriel", "Alice", "Eduardo"}
nombres.remove("Alice") #{"Hannah", "Charlie", "Diana", "Bob", "Fernanda", "Gabriel", "Eduardo"}
nombres.add("Oscar")
print(nombres) # {"Hannah", "Charlie", "Diana", "Bob", "Fernanda", "Gabriel", "Oscar", "Eduardo"}

nombres.discard("Hannah")
print(nombres) #{"Charlie", "Diana", "Bob", "Fernanda", "Gabriel", "Oscar", "Eduardo"}

#actualizando set() para fusionar elementos de  otro set dentro de otro
nombres_nuevos = {"Daniela", "Xavier"}
print(nombres_nuevos)
nombres_nuevos.update(nombres)
print(nombres_nuevos)
````

### `{}.union({})` - `{} | {} `- union de sets

    ```python
    {1,2,4}.union({4,5,6}) # -> {1, 2, 3, 4, 5, 6}
    ```

- funciona por el operador `builtin` dentro de la instancia de un set 
    ```python
    {45, 15, 90} | {"Alice", "hisory",90} #-> {45, 15, 90, "Alice", "hisory",90}
    ```

````python
marcas_coches = {"Toyota", "BMW", "Ford"}
marcas_empresas = {"Siemens", "Bosh"}
marcas_empresas.union(marcas_coches) #devuelve pero no modifica el original y remueve duplicados
print(marcas_empresas)
print(marcas_empresas | marcas_coches)
````

## `dict()` `a={name:"Marcy"}`– Diccionarios
- estructura de datos almacenada en conjuntos de **llave-valor** `key-value pairs`
- Similares a los JSON `{"name": "Marcy"}`
- Se usan para acceder de manera rapida a informacion asociada a cierto elemento
- checar si una llave existe dentro de un `dic`cionario es una operacion muy rapida a comparacion de `list`
- no se pueden acceder por indice

- Se asocian con *keys* [hashables](https://docs.python.org/3/glossary.html#term-hashable) hacia *values*,
  - sus ***keys*** deben ser unicas a lo largo del `dict()` y por ende hasaheables y pueden incluir:
    - `numbers` /**float** / **int**
    - `str()`
    - `tuple()`
  - Su orden de "key-values" esta garantizado en el orden en que fuero insertadas
  - `values` -> pueden ser de cualquier tipo y esturctura de datos
- `dict()` ocupan más memoria pero son bastante más rapidos en terminos de busqueda de datos dentro de alguna estructura comparados con `list`
  - su caso de uso es acceso y actualizacion frecuentes en grandes colecciones de datos

````python
print(type({})) #checamos el tipo de dato: dict

# diferencia entre set y dict es un simbolo , 
print({1: "one"})

nums = {
    "one": 1, 
    "two": 2, 
    "three": 3
}

print(nums)
print(nums["one"])
try: 
    print(nums["four"])
except KeyError as e:
    print(f"Key {e} no existe en el diccionario")

````

### `{}.get(<llave>)` `{}.get(<llave>, <valor_por_defecto>)` – Retornando el valor de una llave
- podemos retornanr el valor de una llave  dentro de un `dic`cionario `{}.get(<llave>)`
- si no existe el **key** o llave, no retornará nada, por lo que el metodo `{}.get(<llave>, <valor_por_defecto>)`, nos permite pasar como segundo parametro un valor por defecto de retorno por si no encuentra la llave

````python
motos = {
    "marca": "Honda"
}

print(motos.get("marca"))
print(motos.get("modelo"))#retorna None
print(motos.get("modelo", "No current key model assigned"))#retorna nuestra cadena por defecto o valor por defecto
````

### `{}.udpate(<otro_dict>)` `{}.update({"key": "value"})` – Agregando y Modificando un diccionario
- dado que si se puede modificar podemos utilizar el metodo u operador de asignacion 


````python
print(motos)
motos["modelo"] = "Monkey (Z50)" # agregando nuevos key-value pairs a un dict
print(motos)

motos_detail = {
    "precio": 20,
    "color": "green"
}

motos.update(motos_detail)
print(motos)#agrega los key-values, si ya existen retorna el nuevo valor actualizado
motos.update({"modelo": "CD70"})
print(motos)
````

### `{}.keys()` `{}.values()` `.items()` – retornando solo llaves o valores mediante el `builtin method`
- nos permite retornar todas las `keys` o `values` de un `dict`
- `{}.items()` -> retorna una lista de tuplas `tuple` con los `key.value pairs` 

````python
print(motos)
print(motos.keys())    # dict_keys  -> solo llaves
print(motos.values())  # dict_values -> solo valores
print(motos.items())   # dict_items  -> vista de tuplas, iterable pero no indexable (key, value)

# para acceder por indice, debes convertir a list primero
lista_elementos = list(motos.items())  # -> lista de tuplas
print(lista_elementos)          #dict_items([("marca", "Honda"), ("modelo", "CD70"), ("precio", 20), ("color", "green")])
print(lista_elementos[0])      # -> primera tupla ("marca", "Honda")
print(lista_elementos[0][0])   # -> key   "marca"
print(lista_elementos[0][1])   # -> value "Honda"
````

### Comparativos `4 > 5` `is` `==`
- `is` – compara identidad del objeto, pregunta si mabas variables apuntan al mismo objeto en memoria
    ```python
    a = [1, 2]
    b = a
    c = [1, 2]
    print(a is b) #true
    print(a is c) #false - checa ubicacion del objeto en memoria
    ```
- `==` – compara valor por valor en la tabla de valores de ***ASCII*** y tambien si tienen el mismo contenido 

````python
print(4 > 5) # false
print(4 >= 5) # false
print(4 < 5) # true
print(4 == 5) # false
print("bat" ==  "cat") # false
print([1,2, 3] ==  [1,2, 3]) # false
print([0,30, 3] ==  [1,2, 3]) # false
a = [1,2, 3]
b = [1,2, 3]
print(a == b) # true

# is -> identity keyword 
x = None
print(f"{x is None}") #true
print(f"{x is a}") #false

a1 = [1, 2]
b1 = a1
c = [1, 2]
print(a1 is b1) #true
print(a1 is c) #false - checa ubicacion del objeto en memoria
````

## `loop over a dict()` iterando sobre un `dict()` – **tuple unpacking** y `enumerate( <dict> / <list> )`
- `enumerate(<dict> / <str>)` – puede recibir un parametro de tipo `list()`, `tuple()`, devuelve un **dict()** con *keys* numericos de un indice empezando por `0`

````python
hex_colors = {
    "Red": "#FF",
    "Green": "#008",
}

for hex_color in hex_colors:
    print(hex_color)#imprime el key solamente

for hex_color in hex_colors.items():
    print(type(hex_color), hex_color) # retorna tuples de los key-value pairs
    #usar desempaquetado de tuples()
    
for label, hex in hex_colors.items():
    print(label, hex)# retorna en label el key y en hex el value | tuple() – unpacking

#usar el metodo built-in enumerate 
for i in enumerate(hex_colors.values()):
    print(i)
    
for i, key_value_pair in enumerate(hex_colors.items()):
    print(i, key_value_pair)

for obj in hex_colors.items():
    print(obj)
    print(obj[0])
    print(obj[1])
````

- Pasos:
    1) importar la bibliota `requests`
    2) definir el endpoint que ocuparemos para realizar requests

````python
import requests #importar biblioteca requests
import json

API_URL = "https://api.github.com/search/repositories" #definir endpoint
headers = {"Accept": "application/vnd.github+json", "X-GitHub-Api-Version": "2026-03-10"} #definir headers
url_params = {
    "q": "stars:>5000", #buscar repos con extrellas mayores a 5000, parametros especificos para esta API, no son propios de python
    "type": "Repositories", #query param pra obtener innformacion de repos
    "sort": "stars", #filtrar por numeor de stars en el repo
    "order": "desc", #query param - orden descendente
    "per_page": 8#traer dos repos

} # Definir query parameters

def repos_with_most_starts():
    response = requests.get(API_URL, headers= headers, params = url_params )

    print(response.text)
    print(
        
        json.dumps(response.text)# convertir respuesta de texto a json
    )
    try: 
        with open("./app/response.json","w") as text_file: #open(<nombre_archivo.file>,<w>-operacion de escritura)
            text_file.write(response.text)
    except OSError as e:
        print(e)


if __name__ == "__main__": #ejecutar funcion solo si es ejecutado el archivo directamente y no por importacion como biblioteca
    repos_with_most_starts()

    pass #no hacer nada
````
