### Funciones con valores por defecto
- Los argumentos de la funcion que tienen valores por defeco deben de ser  escritos al utlimo dentro de los parentesis, de otra forma obtendremos un `syntax error`
- (❌ ♻️) No se debe prestablecer como valor por defecto una `list` vacia, ya que los parametros asignados a las funciones se instancia una vez, y es al inicio de la ejecucion de la función, por lo que no podrás usar nuevamente la funcion sin que se asignen una y los parametros que le pases, es decir permaneceran dentro de la funcion y ya no poadrá ser reusable

````python
def saludo (nombre, saludo= "Hola "):
    return print(f"{saludo} {nombre}")

saludo("Juan")

def agregar_numero(a, b=[]): #instancia los paramtros 1 sola vez al crearse funcion
    b.append(a)
    return print(b)

agregar_numero(1)
agregar_numero(5) #arrastra el primer valor 😭
agregar_numero(6) #arrastra el primer y segundo valor 😭 y no devuelve una nueva lista, no lo convierte en algo reutilizable ♻️

# lo que se debe hacer es asignarle un valor None
def agregar_numero_mejorado(a, b=None):
    if b ==None:
        b = []
    b.append(a)
    return print(b)

agregar_numero_mejorado(2) # ya no arrastra los valores de inicio
agregar_numero_mejorado(3) # ya no arrastra los valores de inicio
agregar_numero_mejorado(88) # ya no arrastra los valores de inicio
````

### Ordenando listas
- **Ordenar de menor a mayor** – puedes hacerlo de 2 maneras:
  - `sorted()` -> no modifica el arreglo original
  - `[34,14].sort()` -> SI modifica el arreglo original -> `[14, 34]`
- **Revertir el orden actual**:
  - `[19, 201, 2].revert()` -> retorna `[2, 201, 19]`

````python
numeros_suerte = [45, 888, 24]

print(
    f"{sorted(numeros_suerte)} 'sorted' no modifica el arreglo original \n arreglo original= {numeros_suerte}"
)

numeros_suerte.sort() #este si modifica el nuevo arreglo
print(numeros_suerte)


nuevos_numeros_suerte = [204, 305, 40]
print(nuevos_numeros_suerte)
nuevos_numeros_suerte.reverse()
print(nuevos_numeros_suerte)
````

### `"<value>" in []` Busqueda dentro de `list`
- existen diversas formas, aunque es una operacion muy lenta y puede tomar bastante tiempo
  - **saber si existe un elemento en esa posicion**:
    ```python
    names = ["Alice", "Bob", "Charlie", "Diana", "Eduardo", "Fernanda", "Gabriel", "Hannah"] 
    "Gabriel" in names #retorna-> True
    "Nina" in names #retorna-> False
    ```
    - Returna `True` o `False` 
- `[].index(<value>)` – retorna si es verdad el indice, si es faslo `-1`
$x=\sqrt{2}=x_{m}^{3}$

````python
names = ["Alice", "Bob", "Charlie", "Diana", "Eduardo", "Fernanda", "Gabriel", "Hannah"] 
print(names.index("Bob")) # devuelve el indice donde encunetra ese elemento
print("Gabriel" in names)

print(names.count("Bob")) # retorna el numero de veces que aparece "Bob" en la lista
````

### Removiendo elementos de una lista
- `[].remove(<elemento_del_arreglo_a_remover>)`solo remueve el primer elemento de la lista si existen dos instancias del mismo elemento
- si no existe el elemento que no existe dentro de la lista retorna un error 
- puedes usar el metodo `pop` para remover el ultimo elemento de la lista -> y retornara el elemento removido de la lista
  - `[].pop(<indice>)` -> utiliza el indice y la lista para remover un elemento en especifico
  - `[].pop()` -> ultimo elemento de la lista -> y retornara el elemento removido de la lista

````python
names_duplicate = ["Alice", "Bob", "Charlie", "Diana", "Eduardo", "Fernanda", "Gabriel", "Hannah", "Bob", "Gabriel"]
print("Gabriel" in names_duplicate)

print(names_duplicate.index("Gabriel"))
````
