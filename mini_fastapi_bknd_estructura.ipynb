# Estructura basica de un mini proyecto `fastAPI` 

```text
mini_fastapi_project
|
|--- venv/
|--- app/
|     \--- main.py
\--- requirements.txt
```

- Donde:
  - `venv/` -> entorno virtual
  - `app/main.py` -> entorno virtual
## Creando el proyecto:

- Crea la carpeta del proyecto, esta vez podemos usar `bash` para crear nuestra carpeta / directorio raiz del proyecto:
```bash
mkdir mini_fastapi_project # creando el directorio raiz del proyecto
cd mini_fastapi_project # cambiarnos/ingresamos al directorio
python3 -m venv venv # ejecutar python3 con la opcion de -m "modulos .py" del cual buscara alguno que coincida con venv y el ultimo venv lo toma como argumento del nombre de la carpeta
touch requirements.txt # crear archivo en el que documentaremos las dependencias que instalaremos
mkdir app # crear carpeta app
touch app/main.py # crear archivo main.py dentro de carpeta app/

source venv/bin/activate # activar el entorno venv de python
```

## Instalar `FastAPI` y `Uvicorn`

```bash
pip install fastapi uvicorn
```
- `fastapi` -> **framework** para crear la API, similar a `express.js`
- `uvicorn` -> ***Servidor web*** en python para gestionar operaciones asincornas `asyn I/O`) **ASGI - Asynchronous Server Gateway Interface** 
  - Funcionalidad similar a `nodemon` en algunas cosas como reiniciar el server automaticamente cuando cmabia el codigo
  - permite programacion asincrona para manejar muchas conexiones al mismo tiempo:
    ```text
    - WebSockets
    - peticiones simultaneas
    - APIs rapidas
    - streaming de datos
    - app en tiempo real (chat, notificaciones)
    ```
  ### comparativo `uvicorn` vs `node.js`


  |Concepto|Python|JavaScript|
  |-|-|-|
  |Framework web| FastAPI | JavaScript |
  | Interfaz web | ASGI | API HTTP Node |
  |Servidor | Uvicorn | Runtime Node |
  | Hot Reload | `uvicorn --reload` | `nodemon` |
